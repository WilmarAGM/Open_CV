```python
from PIL import Image, ImageDraw, ImageFont
```

```python
image = Image.open('hViYL.png')
image = image.convert('RGB')
display(image)
print(image.width, image.height)
```

```python
factores = [0.1,0.5,0.9]

images = []
for i in factores:
    picture_new = Image.new('RGB', (image.width, image.height))
    for x in range(image.width):
        for y in range(image.height):
            r,g,b = image.getpixel((x,y))
            r = int(i*r)
            picture_new.putpixel((x,y), (r,g,b))
    picture_text = ImageDraw.Draw(picture_new)
    font = ImageFont.truetype("arial.ttf", 40)
    texto = 'chanel 0 intensity {}'.format(i)
    picture_text.text((100,image.height-(40+5)), texto, font= font, fill=(255,255,255,0))
    images.append(picture_new)


for i in factores:
    picture_new = Image.new('RGB', (image.width, image.height))
    for x in range(image.width):
        for y in range(image.height):
            r,g,b = image.getpixel((x,y))
            g = int(i*g)
            picture_new.putpixel((x,y), (r,g,b))
    picture_text = ImageDraw.Draw(picture_new)
    font = ImageFont.truetype("arial.ttf", 40)
    texto = 'chanel 1 intensity {}'.format(i)
    picture_text.text((100,image.height-(40+5)), texto, font= font, fill=(255,255,255,0))
    images.append(picture_new)      

for i in factores:
    picture_new = Image.new('RGB', (image.width, image.height))
    for x in range(image.width):
        for y in range(image.height):
            r,g,b = image.getpixel((x,y))
            b = int(i*b)
            picture_new.putpixel((x,y), (r,g,b))
    picture_text = ImageDraw.Draw(picture_new)
    font = ImageFont.truetype("arial.ttf", 40)
    texto = 'chanel 2 intensity {}'.format(i)
    picture_text.text((100,image.height-(40+5)), texto, font= font, fill=(255,255,255,0))
    images.append(picture_new) 


first_image = images[0]

contact_sheet = Image.new(first_image.mode, (first_image.width*3,first_image.height*3))
x=0
y=0

for img in images:
    # Lets paste the current image into the contact sheet
    contact_sheet.paste(img, (x, y) )
    # Now we update our X position. If it is going to be the width of the image, then we set it to 0
    # and update Y as well to point to the next "line" of the contact sheet.
    if x+first_image.width == contact_sheet.width:
        x=0
        y=y+first_image.height
    else:
        x=x+first_image.width

# resize and display the contact sheet
contact_sheet = contact_sheet.resize((int(contact_sheet.width/2),int(contact_sheet.height/2) ))
display(contact_sheet)
```
